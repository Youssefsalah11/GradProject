# Our First Model
## Dataset

```python
import pandas as pd
from torchvision.io import read_image
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import ToTensor

class LaneDetectionDataset(Dataset):
    def __init__(self, annotations_file):
        # Load the CSV file with annotations
        self.annotations = pd.read_csv(annotations_file)
        
    def __len__(self):
        # The length of the dataset is the number of rows in the CSV
        return len(self.annotations)

    def __getitem__(self, idx):
        # Get the row for the current index
        row = self.annotations.iloc[idx]
        
        # Load the image at the image_path
        image = read_image(row['File Path'])

        # Extract all lane coordinates from the row, convert them to a float tensor
        # Assuming all remaining columns are lane coordinates
        lane_coords = row[1:].values.astype('float32')
        lane_coords = torch.tensor(lane_coords, dtype=torch.float)

        return image, lane_coords

```

### Model

```python
import torch
import torch.nn as nn
import timm

class LaneDetectionModel(nn.Module):
    def __init__(self, num_lanes, num_x_coordinates_per_lane):
        super(LaneDetectionModel, self).__init__()

        # Load pre-trained ConvNeXt-Base model
        convnext_base = timm.create_model('convnext_base', pretrained=True)
        self.features = nn.Sequential(*list(convnext_base.children())[:-2])  # Remove the head

        # Assuming the feature size of ConvNeXt-Base is 1024 for the penultimate layer
        self.global_avg_pool = nn.AdaptiveAvgPool2d((1, 1))
        self.fc = nn.Linear(1024, num_lanes * num_x_coordinates_per_lane)

        self.num_lanes = num_lanes
        self.num_x_coordinates_per_lane = num_x_coordinates_per_lane

    def forward(self, x):
        # Extract features using ConvNeXt-Base
        x = self.features(x)
        x = self.global_avg_pool(x)
        x = torch.flatten(x, 1)  # Flatten the features into a 1D vector

        # Perform regression to get x-coordinates
        x = self.fc(x)

        x = x.view(-1, self.num_lanes, self.num_x_coordinates_per_lane)
        return x
```

### Loss Fucntion

```python
import torch
import torch.nn as nn

class CustomLoss(nn.Module):
    def __init__(self):
        super(CustomLoss, self).__init__()

    def forward(self, predicted_x_coords, target_x_coords):
        # Mask indicating valid (non-missing) values in target_x_coords
        valid_mask = (target_x_coords != -1).float()

        # Calculate the mean squared error (MSE) loss only for valid values
        loss = torch.sum(valid_mask * (predicted_x_coords - target_x_coords) ** 2) / (valid_mask.sum() + 1e-9)

        return loss
```

### Training

```python
def reshape_lane_coords(lane_coords, num_lanes=4, num_x_coordinates_per_lane=19):

    image_width=1640
    
    #normalize the coords
    lane_coords=lane_coords/image_width
    batch_size = lane_coords.size(0)
    reshaped_lane_coords = lane_coords.view(batch_size, num_lanes, num_x_coordinates_per_lane)
    
    return reshaped_lane_coords
```

```python
import torch
import torch.optim as optim
from torch.utils.data import DataLoader



batch_size = 4
num_epochs = 10

# Dataset and model instances
dataset = LaneDetectionDataset(annotations_file='annotations.csv')
model = LaneDetectionModel(num_lanes=4, num_x_coordinates_per_lane=19)

data_loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)
loss_function = CustomLoss()
learning_rate = 0.01
optimizer = optim.SGD(model.parameters(), lr=learning_rate)



for epoch in range(num_epochs):
    
    epoch_loss = 0.0
    
    for images, lane_coords in data_loader:

        images = images.to(torch.float32) 
        
        # reshape lane_coords
        lane_coords = reshape_lane_coords(lane_coords, num_lanes=4)
        
        # forward pass
        predicted_lane_coords = model(images)

        # compute loss
        loss = loss_function(predicted_lane_coords, lane_coords)
        
        # backward pass and optimization
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        # logging the loss
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item()}')
        epoch_loss += loss.item()
        
    # Calculate the average loss for the epoch
    epoch_loss /= len(data_loader)
    print(f'Epoch [{epoch+1}/{num_epochs}], Average Loss: {epoch_loss}')
```

```python
import cv2
from torchvision import transforms
import numpy as np

def draw_lanes_on_image(image_path,lines):
    # Read the image
    image = cv2.imread(image_path)

    y_coords = [i for i in range(590,400,-10)]
    for idx,line in enumerate(lines):
        x_coords=line
        # draw the lane on the image
        for i in range(0, 18):
            if x_coords[i]<0:
                continue
            if idx==0:
                cv2.circle(image, (int(x_coords[i]), int(y_coords[i])), 5, (255,0, 0), -1)
            elif idx==1:
                cv2.circle(image, (int(x_coords[i]), int(y_coords[i])), 5, (0,255, 0), -1)
            elif idx==2:
                cv2.circle(image, (int(x_coords[i]), int(y_coords[i])), 5, (0,0, 255), -1)
            else:
                cv2.circle(image, (int(x_coords[i]), int(y_coords[i])), 5, (255,255, 0), -1)
    output_path = "annotated_" + image_path.split('/')[-1]
    cv2.imwrite(output_path, image)
    cv2.imshow('Annotated Image', image)
    cv2.waitKey(0)
    cv2.destroyWindow('Annotated Image')


```

```python
# path to model and path to image to test
MODEL_PATH = 'first_model1.pth'
IMAGE_PATH = 'trainingSet/driver_161_90frame/06030904_0770.MP4/04860.jpg'

# Preprocess the image
def preprocess_image(image_path):
    image = cv2.imread(image_path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert to RGB
    # Perform any other preprocessing steps your model requires:
    # resizing, normalization, etc.
    transform = transforms.Compose([
        transforms.ToTensor(),
        # Add other transformations as per your model's requirements
    ])
    image_tensor = transform(image).unsqueeze(0)  # Add batch dimension
    return image_tensor, image
    
# Assuming preds is a flat list or a 1D tensor of normalized x-coordinates
def postprocess_predictions(preds, width):
    # Convert the predictions from tensor to numpy array if not already
    preds = preds.squeeze(0).cpu().numpy() 


    # Scale the predictions to the image dimensions
    x_coords = preds * width
    
    return x_coords

#load model
model = LaneDetectionModel(num_lanes=4, num_x_coordinates_per_lane=19)
model.load_state_dict(torch.load(MODEL_PATH))
model.eval() 

#test image
image_tensor,image=preprocess_image(IMAGE_PATH)
with torch.no_grad():
     preds = model(image_tensor)


lines_lanes=postprocess_predictions(preds, 1640)
draw_lanes_on_image(IMAGE_PATH,lines_lanes)
```
